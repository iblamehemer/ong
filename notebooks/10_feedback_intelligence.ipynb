## 2. Generate Simulated Feedback Dataset

```python
import random, csv
from pathlib import Path

MODULES = ['Logo & Design Studio','Creative Content Hub','Campaign Analytics',
           'Multilingual Studio','Overall Brand Kit']
COMPANIES = ['NovaTech','LuxeMode','GreenSeed','HealthAI','RetailBoost']
INDUSTRIES = ['Technology / Software','Fashion / Apparel','Sustainability / Green Tech',
              'Healthcare','Retail / E-commerce']

random.seed(42)
rows = []
for i in range(100):
    rating   = random.choices([1,2,3,4,5], weights=[5,8,15,42,30])[0]
    ci       = random.randint(0,4)
    rows.append({
        'timestamp':  f'2025-0{random.randint(1,9)}-{random.randint(10,28)}T{random.randint(9,21)}:00:00',
        'session_id': f'sess_{i:04d}',
        'company':    COMPANIES[ci],
        'industry':   INDUSTRIES[ci],
        'module':     random.choice(MODULES),
        'rating':     rating,
        'sentiment':  'positive' if rating>=4 else 'neutral' if rating==3 else 'negative',
        'comment':    random.choice(['Great output!','Needs improvement','Very accurate','Loved the palette','Taglines were perfect','']),
        'preferred':  '',
    })

fb_path = Path('simulated_feedback.csv')
with open(fb_path,'w',newline='') as f:
    writer = csv.DictWriter(f, fieldnames=rows[0].keys())
    writer.writeheader()
    writer.writerows(rows)

df = pd.read_csv(fb_path)
print(f'Simulated feedback: {len(df)} records')
print(df[['module','rating','sentiment']].head(10))
```

## 3. Summary Statistics

```python
summary = get_summary(df)
if summary:
    print(f'Total submissions: {summary["total_submissions"]}')
    print(f'Average rating:    {summary["avg_rating"]}')
    print(f'Positive:          {summary["positive_pct"]}%')
    print(f'Neutral:           {summary["neutral_pct"]}%')
    print(f'Negative:          {summary["negative_pct"]}%')
    print(f'\nBy module:')
    for mod, avg in summary['by_module'].items():
        bar = '█' * int(avg*4)
        print(f'  {mod:30s}: {avg:.2f} {bar}')
```

## 4. Plotly Feedback Dashboard

```python
fig1 = px.bar(df, x='module', y='rating', color='sentiment',
             color_discrete_map={'positive':'#3ECFB2','neutral':'#C9A84C','negative':'#E05A5A'},
             title='Ratings by Module', barmode='group')
fig1.update_layout(paper_bgcolor='#141518',plot_bgcolor='#141518',font_color='#F0EDE8',
                   title_font_color='#C9A84C')
fig1.show()

fig2 = px.pie(df['sentiment'].value_counts().reset_index(),
              names='sentiment', values='count',
              color='sentiment',
              color_discrete_map={'positive':'#3ECFB2','neutral':'#C9A84C','negative':'#E05A5A'},
              hole=0.55, title='Sentiment Distribution')
fig2.update_layout(paper_bgcolor='#141518',font_color='#F0EDE8',title_font_color='#C9A84C')
fig2.show()
```

## 5. Model Refinement Strategy

```python
# Identify which modules need improvement
module_avg = df.groupby('module')['rating'].mean().sort_values()
print('Module ratings (lowest first):')
for mod, avg in module_avg.items():
    status = '🔴 Needs work' if avg < 3.5 else '🟡 Good' if avg < 4.2 else '🟢 Excellent'
    print(f'  {mod:30s}: {avg:.2f} {status}')

print('\nRefinement actions:')
print('  • Modules rated < 3.5 → retrain with negative feedback as hard negatives')
print('  • Collect preferred_alternative text → use as new training hints')
print('  • Weekly feedback aggregation → prompt engineering updates')
```
