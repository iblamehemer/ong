## 1. Logo Engine Overview

Since no logo image dataset was uploaded, we implement a **rule-based SVG logo generator** that:
- Extracts initials from the company name
- Applies the brand colour palette
- Generates 5 distinct geometric styles
- Exports to SVG and PNG

```python
from src.logo_engine import generate_all_logos, generate_svg_logo, _get_initials, svg_to_png_bytes
from src.palette_engine import generate_palette

# Test initials extraction
test_companies = ['NovaTech Solutions', 'Apple', 'The Coca-Cola Company', 'Zara', 'BMW']
for c in test_companies:
    print(f'{c:30s} → {_get_initials(c)}')
```

## 2. Generate All 5 Logo Styles

```python
# Build a test palette
palette = generate_palette('Technology / Software', 'Minimalist')
print('Palette colors:', [v['hex'] for v in palette.values()])

# Generate all 5 logo concepts
logos = generate_all_logos('NovaTech', palette)
for logo in logos:
    print(f"Style {logo['index']+1}: {logo['style']:20s} — {len(logo['svg'])} chars SVG")
```

## 3. Visualise Each SVG Style

```python
# Display each SVG concept
for logo in logos:
    print(f"\n── {logo['style']} ──")
    print(f"Description: {logo['description']}")
    display(SVG(logo['svg']))
```

## 4. Test Across Industries & Personalities

```python
test_cases = [
    ('LuxBrand',  'Fashion / Apparel',        'Luxury'),
    ('EcoGrow',   'Sustainability / Green Tech','Vibrant'),
    ('FinServe',  'Finance',                    'Professional'),
    ('FoodieApp', 'Food & Beverage',            'Playful'),
]

for company, industry, personality in test_cases:
    pal = generate_palette(industry, personality)
    logos = generate_all_logos(company, pal)
    print(f'{company:12s} ({personality:12s}): {len(logos)} logos, primary={list(pal.values())[0]["hex"]}')
```

## 5. PNG Export via Pillow

```python
# Export first logo style as PNG
palette = generate_palette('Technology / Software', 'Minimalist')
logos   = generate_all_logos('NovaTech', palette)
png     = svg_to_png_bytes(logos[0]['svg'], size=300)

print(f'PNG size: {len(png)//1024} KB')
assert png[:8] == b'\x89PNG\r\n\x1a\n' or len(png) > 0

img = Image.open(io.BytesIO(png))
print(f'Image dimensions: {img.size}')
plt.figure(figsize=(4,4))
plt.imshow(img)
plt.axis('off')
plt.title('Logo PNG Export', color='white')
plt.tight_layout()
plt.show()
print('✅ PNG export successful')
```

## 6. Logo Colour Analysis

```python
from src.palette_engine import hex_to_rgb

for industry in ['Technology / Software', 'Fashion / Apparel', 'Food & Beverage']:
    pal  = generate_palette(industry, 'Minimalist')
    pri  = list(pal.values())[0]
    rgb  = hex_to_rgb(pri['hex'])
    lum  = 0.299*rgb[0] + 0.587*rgb[1] + 0.114*rgb[2]
    tone = 'Dark background' if lum < 128 else 'Light background'
    print(f'{industry:35s}: {pri["hex"]}  luminance={lum:.0f}  → {tone}')
```
