Covers:

```python
import sys
sys.path.insert(0, '..')
import io
from PIL import Image
from IPython.display import display as ipy_display, Image as IPyImage
from src.animation_engine import create_brand_gif
from src.palette_engine import generate_palette
from src.logo_engine import generate_all_logos
print('✅ Libraries loaded')
```

## 1. Build Test Assets

```python
company  = 'NovaTech'
tagline  = 'Engineering the future, one line at a time.'
palette  = generate_palette('Technology / Software', 'Minimalist')
logos    = generate_all_logos(company, palette)
svg      = logos[0]['svg']

print(f'Company:  {company}')
print(f'Tagline:  {tagline}')
print(f'Primary:  {list(palette.values())[0]["hex"]}')
print(f'SVG len:  {len(svg)} chars')
```

## 2. Generate All Three Styles

```python
styles = ['typewriter', 'fade', 'slide']
gifs   = {}

for style in styles:
    gif = create_brand_gif(svg, tagline, palette, company, style, frames=20)
    gifs[style] = gif
    print(f'  {style:12s}: {len(gif)//1024} KB, starts_with_GIF={gif[:3]==b"GIF"}')
```

## 3. Display Animations

```python
for style, gif_bytes in gifs.items():
    print(f'\n── {style.upper()} style ──')
    ipy_display(IPyImage(data=gif_bytes))
```

## 4. Frame Analysis

```python
# Analyse the typewriter GIF frame by frame
gif_bytes = gifs['typewriter']
img = Image.open(io.BytesIO(gif_bytes))

frame_count = 0
try:
    while True:
        frame_count += 1
        img.seek(frame_count)
except EOFError:
    pass

print(f'Frames:      {frame_count}')
print(f'Dimensions:  {img.size}')
print(f'Mode:        {img.mode}')
print(f'Total size:  {len(gif_bytes)//1024} KB')
print(f'Per frame:   {len(gif_bytes)//frame_count//1024} KB avg')
```

## 5. Export to File

```python
import os
os.makedirs('../assets/sample_exports', exist_ok=True)

for style, gif_bytes in gifs.items():
    path = f'../assets/sample_exports/brand_anim_{style}.gif'
    with open(path, 'wb') as f:
        f.write(gif_bytes)
    print(f'✅ Saved: {path} ({len(gif_bytes)//1024} KB)')
```

## 6. Multi-Industry Preview

```python
test_brands = [
    ('LuxeMode',  'Fashion / Apparel',        'Luxury',        'Style is a language. Speak it fluently.'),
    ('GreenSeed', 'Sustainability / Green Tech','Vibrant',       'Grow what matters.'),
    ('HealthAI',  'Healthcare',                'Professional',  'Precision care. Every time.'),
]

for company, industry, pers, tag in test_brands:
    pal = generate_palette(industry, pers)
    svg = generate_all_logos(company, pal)[0]['svg']
    gif = create_brand_gif(svg, tag, pal, company, 'typewriter', 15)
    print(f'{company:12s} ({pers:12s}): {len(gif)//1024} KB GIF generated')
    ipy_display(IPyImage(data=gif))
```
