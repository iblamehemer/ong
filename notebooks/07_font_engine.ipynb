## 5. Coverage Matrix Visualisation

```python
fig, ax = plt.subplots(figsize=(14, 6))
fig.patch.set_facecolor('#0C0D0F')
ax.set_facecolor('#141518')

col_map = {}
for (pers, ind) in FONT_MAP.keys():
    if pers not in col_map:
        col_map[pers] = {}
    col_map[pers][ind] = 1

x_labels = PERSONALITIES
y_labels  = INDUSTRIES

for xi, pers in enumerate(x_labels):
    for yi, ind in enumerate(y_labels):
        key   = (pers, ind)
        color = '#C9A84C' if key in FONT_MAP else '#1C1E22'
        rect  = patches.Rectangle((xi, len(y_labels)-1-yi), 0.9, 0.9, color=color, linewidth=0)
        ax.add_patch(rect)

ax.set_xlim(0, len(x_labels))
ax.set_ylim(0, len(y_labels))
ax.set_xticks([i+0.45 for i in range(len(x_labels))])
ax.set_xticklabels(x_labels, rotation=30, ha='right', color='white', fontsize=9)
ax.set_yticks([i+0.45 for i in range(len(y_labels))])
ax.set_yticklabels(y_labels[::-1], color='white', fontsize=9)
ax.set_title('Font Mapping Coverage Matrix (Gold = explicit, Dark = fallback)', color='white')
plt.tight_layout()
plt.savefig('../assets/sample_exports/font_coverage_matrix.png', dpi=100)
plt.show()
print('✅ Coverage matrix saved')
```
