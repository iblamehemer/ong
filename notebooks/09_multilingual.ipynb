## 1. Translate a Sample Tagline (Fallback Mode)

```python
slogan  = 'Engineering the future, one line at a time.'
company = 'NovaTech'

results = translate_slogan(slogan, company)
results = validate_translations(results)

print(f'Original: "{slogan}"\n')
for lang, data in results.items():
    print(f"{data['flag']}  {lang:10s} ({data['native']:10s}): {data['text']}")
    print(f"   Tone note: {data.get('tone_note','')}")
    print(f"   Source:    {data['source']}\n")
```

## 2. Test Multiple Slogans

```python
test_slogans = [
    ('Just Do It',                         'Nike'),
    ('Think Different',                    'Apple'),
    ('Excellence, redefined.',             'BrandSphere'),
    ('Grow what matters.',                 'GreenSeed'),
    ('Your health. Our mission.',          'HealthPlus'),
]

for slogan, company in test_slogans:
    r = translate_slogan(slogan, company, ['Hindi', 'Spanish', 'French'])
    print(f'\n"{slogan}" ({company})')
    for lang, d in r.items():
        print(f'  {d["flag"]} {lang}: {d["text"]}')
```

## 3. Validate Tone Alignment

```python
# Validate translations with tone notes
for slogan, company in test_slogans[:3]:
    r = translate_slogan(slogan, company)
    r = validate_translations(r)
    print(f'\n"{slogan}"')
    for lang, data in r.items():
        src = '🤖 Gemini' if data['source'] == 'gemini' else '📚 Fallback'
        print(f'  {data["flag"]} {lang:10s} [{src}] — tone: {data.get("tone_note","N/A")[:50]}')
```

## 4. Language Coverage Analysis

```python
import matplotlib.pyplot as plt

langs = list(LANG_META.keys())
flags = [LANG_META[l]['flag'] for l in langs]
speakers = [600, 485, 280, 100, 60, 370, 920]  # approx millions

fig, ax = plt.subplots(figsize=(10, 4))
fig.patch.set_facecolor('#0C0D0F')
ax.set_facecolor('#141518')
bars = ax.bar(langs, speakers, color='#C9A84C', alpha=0.85)
for bar, flag, spk in zip(bars, flags, speakers):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+10, f'{flag}\n{spk}M', 
            ha='center', va='bottom', color='white', fontsize=9)
ax.set_ylabel('Native Speakers (millions)', color='white')
ax.set_title('BrandSphere Supported Language Coverage', color='white')
ax.tick_params(colors='white')
plt.tight_layout()
plt.savefig('../assets/sample_exports/multilingual_coverage.png', dpi=100)
plt.show()
print('✅ Language coverage chart saved')
```
